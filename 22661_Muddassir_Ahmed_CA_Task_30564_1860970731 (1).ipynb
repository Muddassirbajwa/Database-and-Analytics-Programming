The Titanic data set, provided in XML format on Moodle, It provides information about passengers on board the
RMS Titanic, including whether they survived or not.
(a) Create a function to import this XML file. Your function must include appropriate exception handling
clauses covering all possible error conditions and should return the parsed contents of the
XML file

```python
import xml.etree.ElementTree as ET

def parse_xml_file(file_path):
    try:
        # Parse the XML file
        tree = ET.parse(file_path)
        root = tree.getroot()

        # Return the parsed contents
        return root
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
    except ET.ParseError as e:
        print(f"Error parsing XML: {e}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

    # Return None in case of any exception
    return None

# Example usage:
file_path = "dataset.xml"
parsed_xml = parse_xml_file(file_path)
if parsed_xml is not None:
    # You can work with the parsed XML here
    print(ET.tostring(parsed_xml, encoding='utf8').decode('utf8'))
```

b) Using the function you created in a) above, load the XML file. Then use the print() function to display the
Gender and the name for the second, fourth, eighth, sixteenth, and thirty-second passengers.

```python
file_path = "dataset.xml"  # Replace with the path to your XML file
parsed_xml = parse_xml_file(file_path)

if parsed_xml is not None:
    # Extract and print the gender and name of specific passengers
    passengers = parsed_xml.findall("row")

    for index in [1, 3, 7, 15, 31]:
        if index < len(passengers):
            passenger = passengers[index]
            gender = passenger.find("Sex").text
            name = passenger.find("Name").text

            print(f"Passenger {index + 1}:")
            print(f"Gender: {gender}")
            print(f"Name: {name}")
            print()
        else:
            print(f"Passenger {index + 1} does not exist in the XML.")
```

(d) For all passengers where the age less than 30 and Class = 1 or 2, if they survived then extract the fields
below and save them to a CSV file. Note that you should not use Pandas for this task.
• Gender
• Name
• Cabin
• Ticket
• PassengerId
Your code should contain appropriate exception handling clauses and the exported CSV file should contain the
column names

```python
import csv
def save_selected_passengers_to_csv(records):
    # Create or open a CSV file for writing
    with open('selected_passengers.csv', 'w', newline='') as csv_file:
        csv_writer = csv.writer(csv_file)

        # Write the header row to the CSV file
        header = ["Gender", "Name", "Cabin", "Ticket", "PassengerId"]
        csv_writer.writerow(header)

        for record in records:
            age_element = record.find("Age")
            age = float(age_element.text) if age_element is not None and age_element.text is not None else None
            pclass = int(record.find("Pclass").text)
            survived = int(record.find("Survived").text)

            if age is not None and age < 30 and pclass in [1, 2] and survived == 1:
                gender = record.find("Sex").text
                name = record.find("Name").text
                cabin = record.find("Cabin").text
                ticket = record.find("Ticket").text
                passenger_id = record.find("PassengerId").text

                csv_writer.writerow([gender, name, cabin, ticket, passenger_id])

# Load the XML file
file_path = "dataset.xml"  # Replace with the path to your XML file
parsed_xml = parse_xml_file(file_path)

if parsed_xml is not None:
    # Extract and save selected passengers to a CSV file
    passengers = parsed_xml.findall("row")
    save_selected_passengers_to_csv(passengers)

print("Selected passengers have been saved to 'selected_passengers.csv'.")
```
